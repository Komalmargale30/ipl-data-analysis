```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
matches = pd.read_csv('match_details_2025.csv')

matches
```

```python
# Load datasets
batsman = pd.read_csv('ipl_batsman.csv')

batsman
```

```python
# Preview
print(matches.head())
print(batsman.head())
```

```python
# Info
print(matches.info())
print(batsman.info())
```

```python
# Remove duplicates
matches.drop_duplicates(inplace=True)
batsman.drop_duplicates(inplace=True)
```

```python
# Handle missing values
matches.fillna(method='ffill', inplace=True)
batsman.fillna(0, inplace=True)
```

```python
# Standardize column names
matches.columns = matches.columns.str.lower().str.replace(" ", "_")
batsman.columns = batsman.columns.str.lower().str.replace(" ", "_")
```

```python
# Printing Columns Matches
print(matches.columns)
```

```python
# Printing Columns Matches
print(batsman.columns)
```

```python
# Total Teams
print("Total Matches:", matches.shape[0])
print("Teams:", matches['team_1'].unique())
```

```python
# Most Wining Teams
winner_counts = matches['winner'].value_counts()

plt.figure()
winner_counts.plot(kind='bar')
plt.title("Most Winning Teams")
plt.ylabel("Wins")
plt.show()
```

```python
# Toss Win Match Win
toss_win = len(matches[matches['toss_winner'] == matches['winner']])
toss_lose = len(matches) - toss_win

plt.figure(figsize=(5, 5))
plt.pie([toss_win, toss_lose], labels=['Won', 'Lost'],
        colors=['#2E75B6', '#BDD7EE'], autopct='%1.1f%%', startangle=90)
plt.title('Toss Impact', fontweight='bold')
plt.show()
```

```python
# Venu Wins
venue_wins = matches.groupby('venue')['winner'].count().sort_values(ascending=False)

plt.figure()
venue_wins.head(10).plot(kind='bar')
plt.title("Top Venues by Matches")
plt.show()
```

```python
# Max Runs
matches['target_runs'].max()
```

```python
# Tops Run Scorers
top_runs = batsman.groupby('striker')['runs_scored'].sum().sort_values(ascending=False).head(10)

plt.figure()
top_runs.plot(kind='bar')
plt.title("Top Run Scorers")
plt.show()
```

```python
# Strike Rate
sr = batsman.groupby('striker').agg({
    'runs_scored':'sum',
    'balls_faced':'sum'
})

sr['strike_rate'] = (sr['runs_scored']/sr['balls_faced'])*100
sr = sr[sr['balls_faced'] > 100]  # filter small players

top_sr = sr.sort_values(by='strike_rate', ascending=False).head(10)

print(top_sr)
```

```python
# Most Boundries Hit
batsman['boundaries'] = batsman['fours'] + batsman['sixes']

top_boundaries = batsman.groupby('striker')['boundaries'].sum().sort_values(ascending=False).head(10)

top_boundaries.plot(kind='bar')
plt.title("Most Boundaries Hit")
plt.show()
```

```python

total_matches = pd.concat([matches['team_1'], matches['team_2']]).value_counts()
wins = matches['winner'].value_counts()

win_percentage = (wins / total_matches) * 100
print(win_percentage.sort_values(ascending=False))
```

```python
# Home Ground Wins
home_wins = matches[matches['city'] == matches['winner']]
print("Home Win %:", len(home_wins)/len(matches)*100)
```

```python
# Plauer Sats
player_stats = batsman.groupby('striker').agg({
    'runs_scored':'sum',
    'balls_faced':'sum'
})

player_stats['strike_rate'] = (player_stats['runs_scored'] / player_stats['balls_faced']) * 100

# Consistency = runs per match (approx)
player_stats['consistency'] = player_stats['runs_scored'] / player_stats['balls_faced']

top_players = player_stats.sort_values(by='consistency', ascending=False).head(10)
print(top_players)
```

```python
# Best Batsmam
batsman['impact_score'] = batsman['fours']*4 + batsman['sixes']*6

impact_players = batsman.groupby('striker')['impact_score'].sum().sort_values(ascending=False).head(10)
print(impact_players)
```
